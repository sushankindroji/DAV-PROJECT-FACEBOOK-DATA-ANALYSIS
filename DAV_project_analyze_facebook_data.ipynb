```python
# Now we tenure column convert into year joined for tenure analysis

labels = ['0-1 year','1-2 years','2-3 years','3-4 years','4-5 years','5-6 years','6-7 years','7-8 years','8-9 years']

df['year_group'] = pd.cut(df['tenure'],bins=np.arange(-1,3285+1,365),labels=labels,right=True)
df.head()
```

```python
# year group histograph

plt.figure(figsize=(12,7))
sns.histplot(x='year_group',data=df)
plt.show()
```

```python
# actively user  vs age_group

df['activity_per_day'] = (df['friendships_initiated']/df['tenure']) + (df['likes']/df['tenure']).replace([np.nan],0)

plt.figure(figsize=(12,7))
sns.barplot(x='age_group',y='activity_per_day',data=df)
plt.show()
```

```python
# actively user  vs gender

plt.figure(figsize=(12,7))
sns.barplot(data=df,x='gender',y='activity_per_day')
plt.show()
```

```python
# inactive user analysis

inactive_user = df.query('friend_count == 0 and friendships_initiated == 0 and likes == 0')
sns.histplot(data=inactive_user,x='gender',shrink = 0.8)
plt.show()
```

```python
#usage comparision between web and mobile users

active_user = df.query("friendships_initiated!=0 and likes!=0").reset_index(drop=True)
active_user['use_method'] = (active_user['mobile_likes']>= active_user['www_likes']).replace({True:'mobile',False:'web'})

plt.pie(active_user['use_method'].value_counts(),labels=['mobile','web'],autopct = '%1.1f%%')
plt.show()
```
